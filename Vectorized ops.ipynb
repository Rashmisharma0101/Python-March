```python
df['marks'] = [35, 80, 60]
def grades(x):
    if x >= 75:
        return 'A'
    elif x >= 50:
        return 'B'
    else:
        return 'C'

df['grading'] = df['marks'].apply(grades)
df
```

```python
# Must use np.select (vectorized)
```

```python
conditions = [df['marks'] >= 75, df['marks'] >= 50]
choices = ['A', 'B']

df['grading'] = np.select(conditions, choices, default = 'C']
```

```python
# chaining
df = df.assign(grading = np.select(df['marks'] >= 75, df['marks']>= 50], ['A', 'B'], default = 'C'))
```

```python
# Normalize score
minimum =  df['score'].min()
maximum =  df['score'].max()

df['score_normalized'] = (df['score'] - minimum) / (maximum - minimum)
```
