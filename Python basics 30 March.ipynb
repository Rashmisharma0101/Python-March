```python
# Remove duplicates without changing order

l = [1,2,2,3,4,4]

l1 = []

for x in l:
    if x not in l1:
        l1.append(x)

# complexity n2 for loop, linear seac is also n
```

```python
# Optimized code

s  = set()
l1=  []
for x in l:
    if x not in s:
        l1.append(x)
        s.add(x)
l1
# only searching is in set , that is o(1)
# overall complexcity o(n)
# use a list to maintain order and a set for O(1) lookup, reducing complexity to O(n).
```

```python
d = {}
l = [4,5,6,7,4,5,8]

# step 1: frequency
for num in l:
    d[num] = d.get(num, 0) + 1

# step 2: sort by frequency (descending)
result = sorted(d.keys(), key=lambda x: d[x], reverse=False)

print(result)
```

```python
l = [1,2,3,4,5]
target = 6
seen = set()
pairs = set()

for i in l:
    diff = target - i
    if diff in seen:
        pairs.add(tuple(sorted((i, diff))))
    seen.add(i)

print(list(pairs))
    
```
