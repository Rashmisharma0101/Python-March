```python
# factorial

def factorial(n):
    if n <0 :
        return "not defined"
    result = 1
    for i in range(1, n+ 1):
        result  =  result *i
    return result
factorial(5)
    
```

```python
# fibonnaci
def fib_series(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

fib_series(7)
```

```python
# Product of array except self
l = [1,2,3,4]
result = []
for i in range(len(l)):
    prod = 1
    for j in range(len(l)):
        if i != j:
            prod = prod*l[j]
    result.append(prod)
    

result
```
