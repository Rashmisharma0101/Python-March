```python
# outlier detection
```

```python
q1 = df['purchase_amount'].quantile(0.25)
q3 =  df['purchase_amount'].quantile(0.75)

IQR = q3 - q1

upper = q3 + (1.5*IQR)
lower = q1 - (1.5*IQR)

df['outlier'] = df['purchase_amount'].apply(lambda x : 'outlier' if x > upper or x < lower else 'Not outlier')
df
# This is row-by-row (slow)  apply function
#quantile() automatically handles missing values or use nanpercentile
```

```python
# faster solution - Vectorized Solution

mask = (df["purchase_amount"] > upper) | (df["purchase_amount"] < lower)

df['outlier'] = np.where(mask, 'outlier', 'not outlier')

df
```
