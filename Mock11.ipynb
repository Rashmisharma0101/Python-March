# MINI CASE STUDY: E-commerce (Flipkart Style)

```python

import pandas as pd

df = pd.DataFrame({
    'user_id': [1,1,2,2,3,3,4],
    'product': ['A','B','A','C','B','C','A'],
    'category': ['Electronics','Clothing','Electronics','Home','Clothing','Home','Electronics'],
    'price': [1000,500,1200,700,400,800,1500],
    'quantity': [1,2,1,3,1,2,1],
    'date': ['2024-01-01','2024-01-02','2024-01-01','2024-01-03','2024-01-02','2024-01-03','2024-01-04']
})
```

```python
df
```

```python
# Total Revenue
df['total_revenue'] = df['price']*df['quantity']
df
```

```python
print(df['total_revenue'].sum())
```

```python
df.groupby('category')['total_revenue'].sum().sort_values(ascending = False).head(1)
```

```python
df.groupby('user_id')['total_revenue'].sum().sort_values(ascending = False).head(2)
```

```python
# AOV

AOV= df['total_revenue'].sum() / len(df)
AOV
```

```python
df.groupby('product')['quantity'].sum().sort_values(ascending = False).head(2)
```

```python
df.groupby('date')['total_revenue'].sum()
```

```python
r = df.groupby('user_id')['total_revenue'].sum().reset_index(name = 'tr')
r[r['tr'] > 2000]
```

# HEALTHCARE MINI CASE STUDY (MESSY DATA)

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'patient_id': [1,2,2,3,4,5,6,7,8,9],
    'age': [25, np.nan, 45, 35, 60, 29, np.nan, 40, 50, 30],
    'gender': ['F','M','M','F','F',None,'M','F','F','M'],
    'disease': ['Diabetes','Hypertension','Hypertension','Diabetes','Cancer','Diabetes','Cancer','Hypertension','Diabetes','Cancer'],
    'treatment_cost': [2000,3000,3000,1500,np.nan,1800,5000,3200,2100,4000],
    'visit_date': ['2024-01-01','2024-01-02','2024-01-02','2024-01-03','2024-01-04','2024-01-05','2024-01-06','2024-01-07','2024-01-08','2024-01-09']
})
```

```python
df
```

```python
# Goal: Clean data + generate insights for hospital management as a Healthcare Data Analyst
```

```python
# Remove duplicate patients (same patient_id + visit_date)

df = df.drop_duplicates(subset = ['patient_id', 'visit_date'])
```

```python
df
```

```python
# Handle missing age (fill with median)
df['age'] = df['age'].fillna(df['age'].median())
```

```python
df
```

```python
df['gender'] = df['gender'].fillna('Unknown')
```

```python
df['treatment_cost'] = df['treatment_cost'].fillna(df['treatment_cost'].median())
```

```python
import datetime
df['visit_date'] = pd.to_datetime(df['visit_date'])
```

```python
# Create age group column
# (0-30 = Young, 30-50 = Adult, 50+ = Senior)
def age_grp(x):
    if x > 50:
        return 'Senior'
    elif x >= 30 and x <= 50:
        return 'Adult'
    else:
        return 'Young'

df['age_grp'] = df['age'].map(age_grp)
df
```

# EDA

```python
# Total revenue (hospital earnings)
df['treatment_cost'].sum()
```

```python
# Revenue by disease
df.groupby('disease')['treatment_cost'].sum()
```

```python
# Patient count by disease
df.groupby('disease')['patient_id'].size()
```

```python
# Average treatment cost by age group
df.groupby('age_grp')['treatment_cost'].mean()
```

```python
# Which disease is most expensive per patient?
df.groupby('disease')['treatment_cost'].mean()
```

```python
# Identify high-value patients
r = df.groupby('patient_id')['treatment_cost'].sum().sort_values(ascending = False).reset_index(name = 'counts')
```

```python
r[r['counts']> 4000]
```

```python
#Patients visiting more than once

vis = df.groupby('patient_id')['visit_date'].size()
vis[vis>1]  # no patient in this dataset
```

```python
# Gender-wise spending
df.groupby('gender')['treatment_cost'].sum()
```

```python
# Daily revenue trend
df.groupby('visit_date')['treatment_cost'].sum()
```
